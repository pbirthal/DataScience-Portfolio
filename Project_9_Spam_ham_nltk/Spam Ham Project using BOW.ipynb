### BOW, TFIDF, Machine Learning Algorithms 

```python
import pandas as pd 
import re 
import gensim
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
messages = pd.read_csv('data\SMSSpamCollection', sep='\t', names = ['labels', 'message'])
messages
```

```python
ps= PorterStemmer()
```

```python
corpus = []
for i in range(0, len(messages)): 
    review = re.sub('[^a-zA-Z]', " ", messages['message'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
from sklearn.model_selection import train_test_split
y = pd.get_dummies(messages['labels'], dtype= int)
y = y.iloc[:,0].values
X_train, X_test, y_train, y_test = train_test_split(corpus,y, test_size=0.20)
```

### Using BOW 

```python
from sklearn.feature_extraction.text import CountVectorizer  # BOW < Bag of Words Model >
cv = CountVectorizer(max_features=2500, ngram_range=(1,2))
X_train = cv.fit_transform(X_train).toarray() # Independent feature
X_test = cv.transform(X_test).toarray()
```

```python
cv.vocabulary_
```

## Using Naive Bayes 

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect = MultinomialNB().fit(X_train,y_train)
```

```python
y_pred  = spam_detect.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

### Using TF_IDF Model 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tv = TfidfVectorizer(max_features=2500, ngram_range=(1,2))
X = tv.fit_transform(corpus).toarray()
spam_detect_tfidf = MultinomialNB().fit(X_train,y_train)
y_pred =spam_detect_tfidf.predict(X_test)
accuracy_score(y_test,y_pred)
print(classification_report(y_test, y_pred))
```
